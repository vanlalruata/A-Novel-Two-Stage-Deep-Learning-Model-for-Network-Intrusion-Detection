```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/Maruata/Desktop/A novel IDS/Loss_Comparison.csv")
```

```python
sn.set(font_scale=2)

history_dict = data
LSTM_AE_ids_2017_loss = history_dict['LSTM_AE_CICIDS2017_Loss']
LSTM_AE_ids_2017_val_loss = history_dict['LSTM_AE_CICIDS2017_Val_Loss']
LSTM_AE_ids_2018_loss = history_dict['LSTM_AE_CICIDS2018_Loss']
LSTM_AE_ids_2018_val_loss = history_dict['LSTM_AE_CICIDS2018_Val_Loss']

CNN_ids_2017_loss = history_dict['CNN_CICIDS2017_Loss']
CNN_ids_2017_val_loss = history_dict['CNN_CICIDS2017_Val_Loss']
CNN_ids_2018_loss = history_dict['CNN_CICIDS2018_Loss']
CNN_ids_2018_val_loss = history_dict['CNN_CICIDS2018_Val_Loss']

DNN_ids_2017_loss = history_dict['DNN_CICIDS2017_Loss']
DNN_ids_2017_val_loss = history_dict['DNN_CICIDS2017_Val_Loss']
DNN_ids_2018_loss = history_dict['DNN_CICIDS2018_Loss']
DNN_ids_2018_val_loss = history_dict['DNN_CICIDS2018_Val_Loss']

start_by_epoch = 1
epochs = range(start_by_epoch, len(LSTM_AE_ids_2017_loss) + 1)

fig, ax = plt.subplots(figsize=(15,15))
ax.set_facecolor("white")
plt.plot(epochs, LSTM_AE_ids_2017_loss[start_by_epoch-1:], 'x-', linewidth=2)
plt.plot(epochs, LSTM_AE_ids_2017_val_loss[start_by_epoch-1:], 'x-', linewidth=2)
plt.plot(epochs, LSTM_AE_ids_2018_loss[start_by_epoch-1:], 'x--', linewidth=2)
plt.plot(epochs, LSTM_AE_ids_2018_val_loss[start_by_epoch-1:], 'x--', linewidth=2)

plt.plot(epochs, CNN_ids_2017_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, CNN_ids_2017_val_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, CNN_ids_2018_loss[start_by_epoch-1:], '--', linewidth=2)
plt.plot(epochs, CNN_ids_2018_val_loss[start_by_epoch-1:], '--', linewidth=2)

plt.plot(epochs, DNN_ids_2017_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, DNN_ids_2017_val_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, DNN_ids_2018_loss[start_by_epoch-1:], '--', linewidth=2)
plt.plot(epochs, DNN_ids_2018_val_loss[start_by_epoch-1:], '--', linewidth=2)


plt.title('Comparison of Training and Testing loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.legend(['Loss: LSTM-AE Trained with CICIDS2017', 'Validation Loss: LSTM-AE Trained with CICIDS2017', 
           'Loss: LSTM-AE Trained with CICIDS2018', 'Validation Loss: LSTM-AE Trained with CICIDS2018',
           
           'Loss: CNN Trained with CICIDS2017', 'Validation Loss: CNN Trained with CICIDS2017', 
           'Loss: CNN Trained with CICIDS2018', 'Validation Loss: CNN Trained with CICIDS2018',
           
           'Loss: DNN Trained with CICIDS2017', 'Validation Loss: DNN Trained with CICIDS2017', 
           'Loss: DNN Trained with CICIDS2018', 'Validation Loss: DNN Trained with CICIDS2018',
           ], loc='upper right', facecolor='white', framealpha=1)

plt.grid(axis='y', alpha=.3, color='k', linestyle='-', linewidth=2)
plt.grid(axis='x', alpha=.2, color='k', linestyle='-', linewidth=2)
plt.ylim(0.00000,0.030)

plt.show()

fig.savefig('Training_Validation_Loss.pdf') 
```
