```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/Maruata/Desktop/A novel IDS/Accuracy_Comparison.csv")
```

```python
sn.set(font_scale=2)

history_dict = data
LSTM_AE_ids_2017_accuracy = history_dict['LSTM_AE_CICIDS2017_Accuracy']
LSTM_AE_ids_2017_val_accuracy = history_dict['LSTM_AE_CICIDS2017_Val_Accuracy']
LSTM_AE_ids_2018_accuracy = history_dict['LSTM_AE_CICIDS2018_Accuracy']
LSTM_AE_ids_2018_val_accuracy = history_dict['LSTM_AE_CICIDS2018_Val_Accuracy']

CNN_ids_2017_accuracy = history_dict['CNN_CICIDS2017_Accuracy']
CNN_ids_2017_val_accuracy = history_dict['CNN_CICIDS2017_Val_Accuracy']
CNN_ids_2018_accuracy = history_dict['CNN_CICIDS2018_Accuracy']
CNN_ids_2018_val_accuracy = history_dict['CNN_CICIDS2018_Val_Accuracy']

DNN_ids_2017_accuracy = history_dict['DNN_CICIDS2017_Accuracy']
DNN_ids_2017_val_accuracy = history_dict['DNN_CICIDS2017_Val_Accuracy']
DNN_ids_2018_accuracy = history_dict['DNN_CICIDS2018_Accuracy']
DNN_ids_2018_val_accuracy = history_dict['DNN_CICIDS2018_Val_Accuracy']

start_by_epoch = 1
epochs = range(start_by_epoch, len(LSTM_AE_ids_2017_accuracy) + 1)

fig, ax = plt.subplots(figsize=(15,15))
ax.set_facecolor("white")
plt.plot(epochs, LSTM_AE_ids_2017_accuracy[start_by_epoch-1:], 'x-', linewidth=2)
plt.plot(epochs, LSTM_AE_ids_2017_val_accuracy[start_by_epoch-1:], 'x-', linewidth=2)
plt.plot(epochs, LSTM_AE_ids_2018_accuracy[start_by_epoch-1:], 'x--', linewidth=2)
plt.plot(epochs, LSTM_AE_ids_2018_val_accuracy[start_by_epoch-1:], 'x--', linewidth=2)

plt.plot(epochs, CNN_ids_2017_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, CNN_ids_2017_val_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, CNN_ids_2018_accuracy[start_by_epoch-1:], '--', linewidth=2)
plt.plot(epochs, CNN_ids_2018_val_accuracy[start_by_epoch-1:], '--', linewidth=2)

plt.plot(epochs, DNN_ids_2017_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, DNN_ids_2017_val_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, DNN_ids_2018_accuracy[start_by_epoch-1:], '--', linewidth=2)
plt.plot(epochs, DNN_ids_2018_val_accuracy[start_by_epoch-1:], '--', linewidth=2)


plt.title('Comparison of Training and Testing Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.legend(['Accuracy: LSTM-AE Trained with CICIDS2017', 'Validation Accuracy: LSTM-AE Trained with CICIDS2017', 
           'Accuracy: LSTM-AE Trained with CICIDS2018', 'Validation Accuracy: LSTM-AE Trained with CICIDS2018',
           
           'Accuracy: CNN Trained with CICIDS2017', 'Validation Accuracy: CNN Trained with CICIDS2017', 
           'Accuracy: CNN Trained with CICIDS2018', 'Validation Accuracy: CNN Trained with CICIDS2018',
           
           'Accuracy: DNN Trained with CICIDS2017', 'Validation Accuracy: DNN Trained with CICIDS2017', 
           'Accuracy: DNN Trained with CICIDS2018', 'Validation Accuracy: DNN Trained with CICIDS2018',
           ], loc='lower right', facecolor='white', framealpha=1)

plt.grid(axis='y', alpha=.3, color='k', linestyle='-', linewidth=2)
plt.grid(axis='x', alpha=.2, color='k', linestyle='-', linewidth=2)
plt.ylim(0.96,1.002)

plt.show()

fig.savefig('Training_Validation_Accuracy.pdf') 
```
