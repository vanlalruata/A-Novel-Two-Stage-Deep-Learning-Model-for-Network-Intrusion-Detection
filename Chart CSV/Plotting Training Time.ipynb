```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/Maruata/Desktop/A novel IDS/training_time.csv")
```

```python
sn.set(font_scale=2)

history_dict = data
cicids2017_dnn = history_dict['cicids2017_dnn']
cicids2017_cnn = history_dict['cicids2017_cnn']
cicids2017_lstmae = history_dict['cicids2017_lstmae']

cicids2018_dnn = history_dict['cicids2018_dnn']
cicids2018_cnn = history_dict['cicids2018_cnn']
cicids2018_lstmae = history_dict['cicids2018_lstmae']


start_by_epoch = 1
epochs = range(start_by_epoch, len(cicids2017_dnn) + 1)

fig, ax = plt.subplots(figsize=(15,15))
ax.set_facecolor("white")
plt.plot(epochs, cicids2017_dnn[start_by_epoch-1:], '--', linewidth=2)
plt.plot(epochs, cicids2017_cnn[start_by_epoch-1:], '--', linewidth=2)
plt.plot(epochs, cicids2017_lstmae[start_by_epoch-1:], '-', linewidth=2)

plt.plot(epochs, cicids2018_dnn[start_by_epoch-1:], '--', linewidth=2)
plt.plot(epochs, cicids2018_cnn[start_by_epoch-1:], '--', linewidth=2)
plt.plot(epochs, cicids2018_lstmae[start_by_epoch-1:], '-', linewidth=2)


plt.title('Comparison of Training Time')
plt.xlabel('Epochs')
plt.ylabel('Time in Seconds')

plt.legend(['DNN Trained with CICIDS2017', 'CNN Trained with CICIDS2017', 
           'LSTM-AE Trained with CICIDS2017', 
           
           'DNN Trained with CICIDS2018', 'CNN Trained with CICIDS2018', 
           'LSTM-AE Trained with CICIDS2018'
            
           ], loc='center right', facecolor='white', framealpha=1)

plt.grid(axis='y', alpha=.3, color='k', linestyle='-', linewidth=2)
plt.grid(axis='x', alpha=.2, color='k', linestyle='-', linewidth=2)
plt.ylim(0,600)

plt.show()

fig.savefig('Training_Time.pdf') 
```
